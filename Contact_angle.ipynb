```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# ==============================
# Global style (publication ready)
# ==============================
plt.rcParams.update({
    "font.family": "sans-serif",
    "font.size": 10,
    "axes.linewidth": 1.2,      # axis line thickness
})

# ==============================
# Data
# ==============================
D_pristine = np.array([82.59,79.17,75.99])
D_AHPA = np.array([61.89,60.6,56.19])

L_pristine = np.array([85.14,82.67,82.17])
L_AHPA = np.array([55.83,58.99,56.08])

plt.figure(figsize=(5,5))

# X positions
x1, x2 = 0,1
x3, x4 = 3,4

# ==============================
# Mean and SEM
# ==============================
means = [
    D_pristine.mean(),
    D_AHPA.mean(),
    L_pristine.mean(),
    L_AHPA.mean()
]

sems = [
    D_pristine.std(ddof=1)/np.sqrt(len(D_pristine)),
    D_AHPA.std(ddof=1)/np.sqrt(len(D_AHPA)),
    L_pristine.std(ddof=1)/np.sqrt(len(L_pristine)),
    L_AHPA.std(ddof=1)/np.sqrt(len(L_AHPA))
]

x_positions = [x1, x2, x3, x4]

# ==============================
# Error bars (mean ± SEM)
# ==============================
plt.errorbar(
    x_positions, means, yerr=sems,
    fmt='none',
    ecolor='black',
    elinewidth=1.2,   # thicker error bars
    capsize=4,
    capthick=1.2
)

# ==============================
# Mean markers
# ==============================
plt.scatter(x1, means[0], color="#4C72B0", s=40, zorder=3)
plt.scatter(x2, means[1], color="#4C72B0", s=40, zorder=3)

plt.scatter(x3, means[2], color="#DD8452", s=40, zorder=3)
plt.scatter(x4, means[3], color="#DD8452", s=40, zorder=3)

# Horizontal mean lines
for x, m in zip(x_positions, means):
    plt.plot([x-0.18, x+0.18], [m, m], color='black', linewidth=1.2)

# ==============================
# Axes formatting
# ==============================
plt.xticks(
    [x1,x2,x3,x4],
    ["Bare","AHPA","Bare","AHPA"],
    rotation=30,
    fontsize=11
)

plt.ylabel("Contact angle (deg.)", fontsize=12)

# Ticks inside + style
plt.tick_params(
    axis='both',
    direction='in',     # 🔥 ticks inside
    length=4,
    width=1.2,
    labelsize=11
)

# ==============================
# Legend (center-top, larger)
# ==============================
legend_elements = [
    Patch(facecolor="#4C72B0", edgecolor="black", label="D-AHPA"),
    Patch(facecolor="#DD8452", edgecolor="black", label="L-AHPA")
]

plt.legend(
    handles=legend_elements,
    frameon=False,
    fontsize=11,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.15),  # 🔥 move up & center
    ncol=2
)

# ==============================
# Layout
# ==============================
plt.tight_layout()
plt.show()
```
